### Project: Covid19 Data Analysis

```python
# import the libraries
import numpy as np
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
# import the dataset
dataset = pd.read_csv("covid19_Confirmed_dataset.csv")
dataset.head()
```

```python
dataset.shape
```

```python
# delete the useless columns

df = dataset.drop(["Lat", "Long", "Province/State"], axis=1, inplace = True)
dataset.head()
```

```python
# aggreate the rows by the country
corona_dataset_aggreated = dataset.groupby("Country/Region").sum()
```

```python
corona_dataset_aggreated.sample(5)
```

```python
corona_dataset_aggreated.shape
```

```python
# visualize data related to country
corona_dataset_aggreated.loc["China"]
```

```python
corona_dataset_aggreated.loc["China"].plot()
corona_dataset_aggreated.loc["India"].plot()
corona_dataset_aggreated.loc["Spain"].plot()
corona_dataset_aggreated.loc["Andorra"].plot()
corona_dataset_aggreated.loc["Liberia"].plot()
corona_dataset_aggreated.loc["Greece"].plot()
corona_dataset_aggreated.loc["Canada"].plot()
plt.legend()
```

```python
corona_dataset_aggreated.loc["China"][:3].plot()
```

```python
# calculate the first derivative of the curve
corona_dataset_aggreated.loc["China"].diff()
corona_dataset_aggreated.loc["China"].diff().plot()
```

```python
# maximum infection rate
corona_dataset_aggreated.loc["China"].diff().max()
```

```python
corona_dataset_aggreated.loc["India"].diff().max()
```

```python
countries = list(corona_dataset_aggreated.index)
max_infection_rates = []

for c in countries:
    max_infection_rates.append(corona_dataset_aggreated.loc[c].diff().max())
corona_dataset_aggreated["Max_infection_rates"] = max_infection_rates
```

```python
corona_dataset_aggreated
```

```python
# create a new dataframe
corona_data = pd.DataFrame(corona_dataset_aggreated["Max_infection_rates"])
```

```python
corona_data
```

```python
# importing the happiness dataset 2
happiness_report = pd.read_csv("worldwide_happiness_report.csv")
happiness_report
```

```python
# drop the useless columns
useless_cols = ["Overall rank", "Score", "Generosity", "Perceptions of corruption"]
happiness_report.drop(useless_cols, axis =1, inplace=True)
happiness_report.head()
```

```python
happiness_report.set_index("Country or region", inplace =True)
happiness_report.head()
```

```python
# Join the dataset
corona_data.shape
```

```python
happiness_report.shape
```

```python
data = corona_data.join(happiness_report, how = "inner")
data
```
